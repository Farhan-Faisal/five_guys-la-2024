```python
d_df = pd.DataFrame(banksy["discussion_topic"])
```

```python
def get_info(df, info):
    info_list = []

    for x in df: 
        list_of_reply = df['replies']
        for i in list_of_reply:
            info_list.append(i[info])
    
    return info_list
                            
```

```python
result = get_info(d_df, 'message')
result
```

```python
# Specify the file name
file_name = 'message_list.txt'

# Open a file in write mode
with open(file_name, 'w') as file:
    # Write each item in the list to a new line
    for item in result:
        file.write(f"{item}\n \n")

print(f"List saved to {file_name}")
```
